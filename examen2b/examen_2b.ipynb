| |A|B|C|D|
           |---|---|---|---|---|

2013-01-02|  1.212112| -0.173215|  0.119209| -1.044236|

2013-01-04|  0.721555 |-0.706771 |-1.039575 | 0.271860|

2013-01-06| -0.673690 | 0.113648| -1.478427 | 0.524988|

```python
def random_fechas(numero_fechas):
    fecha_inicio = "2020-02-18"
    anadir_dias = np.arange(0, numero_fechas)
    fechas_indice = []
    for x in range(numero_fechas):
        fecha_random = np.datetime64(fecha_inicio) + anadir_dias[x]
        fechas_indice.append(str(fecha_random))
    return fechas_indice
```

```python
arreglo_e2 = np.random.uniform(-10, 10, size=(6, 4))
dataframe_e2 = pd.DataFrame(arreglo_e2,columns = [
        "A",
        "B",
        "C",
        "D"], index=random_fechas(6))
dataframe_e2
```
