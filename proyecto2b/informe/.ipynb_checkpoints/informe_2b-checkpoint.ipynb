# DESARROLLO

Importar librerias necesarias para la ejecución del código

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
paleta = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
papaleta = ["#ffac81", "#ff928b", "#fec3a6", "#efe9ae", "#cdeac0", "#e0607e"]
```

Se definen las columnas a utilizar, la ruta del archivo y se lee el archivo de la ruta. Se hará uso de los 14848 registros del conjunto de datos extraido.

```python
path_dataset = "L://Familia//Documents//2019B-OCTAVOSEMESTRE//Python//py-diaz-yanangomez-leslie-mishell//proyecto2b//arania_2b//libros.csv"
columnas = ["titulo", 
            "autor", 
            "categoria",
            "publicacion",
            "descargas",
            "paginas"]

dataframe = pd.read_csv(path_dataset,encoding='latin-1', header= None)
dataframe.columns = columnas
dataframe.head(10)
```

## Análisis de datos

### Número de libros por categorias

```python
categoria = dataframe["categoria"].value_counts().head(5)
f,ax = plt.subplots(1,1,figsize=(15,4))
print(categoria)
ax = sns.barplot(categoria.index, categoria,palette=papaleta)
```

**Resultados**: Como se observa en el grafico de barras, la categoria que posee un mayor numero de libros albergados en la pagina es *Autoayuda, Referencia* con un total de 1403 libros y corresponde al 9.44% del total de libros. Es seguido por Religion, Espiritualidad con el 7,82% y Ficción Contemporánea (4,94%). 

### Mes y año donde se publicaron mayor cantidad de libros

```python
anio =  dataframe["publicacion"].value_counts().head(5)
f,ax = plt.subplots(1,1,figsize=(15,4))
print(anio)
ax = sns.barplot(anio.index, anio,palette=paleta)
```

**Resultados**: Como se observa en el grafico de barras, el año y mes donde se han publicado una mayor cantidad de libros es Marzo de 2012 con un total de 598 libros y corresponde al 4,03% del total de libros. Las fechas de publicacion que se encuentran en el top 2 y 3 corresponden a Octubre y Noviembre de 2011 con el 3,70% y 3,48% respectivamente. 

### Libros más descargados

```python
libros_descargados = dataframe[["titulo","descargas"]].sort_values(by=['descargas'], ascending=False).head(7)
f,ax = plt.subplots(1,1,figsize=(15,4)) 
ax = sns.barplot(libros_descargados["titulo"], libros_descargados["descargas"],palette=paleta)
```

**Resultados**: Como se observa en el grafico de barras, el libro con mayor numero de descargas es el Kamasutra Ilustrado con un total de 182719 descargas. Es seguido por 101 Ideas Romanticas con 114978 descargas y Antígona (80923). 

### Descargas promedio por categoria

```python
descargas_categoria = dataframe.groupby("categoria")["descargas"].mean().sort_values(ascending=False).head(5)
print(descargas_categoria)
figura_mes = plt.figure(figsize=(15,8))
ax = figura_mes.add_subplot(1, 1, 1)
ax.plot(descargas_categoria)
ax.set_title('DESCARGAS PROMEDIO POR CATEGORIA')
ax.set_xlabel('Categoria')
ax.set_ylabel('Numero de Descargas Promedio')
```

**Resultado:** Se puede observar que la categoría donde se descargan en promedio mayor cantidad de libros es Clásicos: Teatro correspondiente a 5139 descargas en promedio. Podemos deducir la preferencia por los libros clasicos de teatro. Asimismo, lo sigue los libros pertenecientes a Moda y Belleza con un promedio de 3222 descargas

### Promedio de paginas por libro por categoria

```python
paginas_categoria = dataframe.groupby("categoria")["paginas"].mean().sort_values(ascending=False).head(5)
print(paginas_categoria)
ax = paginas_categoria.plot(kind="barh")
plt.xlabel('Promedio de páginas'); 
plt.ylabel('Categoría'); 
ax.invert_yaxis() 
```

### Autores por Categoría

```python
autores = dataframe.groupby("categoria")["autor"].size().sort_values(ascending=False).head(5)
print(autores)
patches,texts=plt.pie(autores,shadow=True,startangle=90)
plt.pie(autores,shadow=True,labels=autores.index,autopct='%1.1f%%')
plt.legend(patches,autores.index,loc="best")
plt.axis("equal")
plt.title("Numero de autores en cada categoría")
plt.tight_layout()
plt.show()
```

**Resultado:** Autoayuda, Referencia es la categoría que cuenta con mayor variedad de autores (1403) correspondiendo al 30,2% del total de autores registrados.

### Libros por autor

```python
libros_autor = dataframe.groupby("autor")["titulo"].size().sort_values(ascending=False).head(5)
f,ax = plt.subplots(1,1,figsize=(15,4))
ax = sns.barplot(libros_autor.index, libros_autor, palette="rocket")
ax.set_ylabel("Número de libros")
ax.set_xlabel("Autor")
```

**Resultado:** Como se puede observar en el grafico, la mayoria de libros (281) no se conoce su autor original por lo que se ha descrito como Anonimo. La colaboracion entre varios autores se encuentra en segundo lugar con un total de 137 libros registrados y un autor en solitario está en el tercer lugar: Yolanda Pinto con un total de 119 libros.
